# Gradient descent based 3D reconstruction

We are representing the 3D model with a 3D NumPy array with "inverted" masses (filled voxel is 0, empty voxel is 1).

Detectors are respresented by 2D arrays with every bin corresponding to total "inverted" mass crossed in a specific direction (number of empty voxels).

Using "real" and "aprroximated" detector data we can calculate the loss function.

$L_2 = \sum_l (\bar{d}_l-d_l)^2, \quad L_1 = \sum_l |\bar{d}_l-d_l|; \quad d_l = \sum_{i\in l} m_i$, where $d_l$ is a detector for specific direction, $m_i$ is the "inverted" mass in the voxel. 

For loss function sum is over all directions and all voxels, for a specific direction sum is over all crossed voxels.

The fitting of the model is performed by gradient steps, where voxel values are modified by the partial derivative of the loss multiplied by learning rate.

$m^1_i = m^0_i - \lambda \frac{\partial L}{\partial m_i}$

For 2 most popular loss functions relevant for this problem:

$\frac{\partial L_2}{\partial m_i} = 2 \sum_{l\in i} (\bar{d}_l-d_l), \quad \frac{\partial L_1}{\partial m_i} = \sum_{l\in i} 1$

```python
# Simulation parameters
global_start = datetime.now()
thresh = 0.4  # visualisation threshold after simulation
size = 1  # voxel size
delta_alpha = 1.0  # parameter to represent the empty voxels
input_dir = 'data/in' # where to save/read input models&detectors
output_dir = 'data/out' # where to save/read(for visualisation) the output models&detectors
```

## Defining the default scene

```python
det_nx = 3  # number of detectors along axis (NxN)
multi_detect = det_nx - 1  # minimal number of intersecting detectors in the anomalous voxel
scene_shape = (32, 32, 32)
scale = scene_shape[2] / 20

angle_bin = 3 * scene_shape[2]  # binning of the angle along axis
# min angle bin number should be 2*np.sqrt(2)*h -- furthest voxel angular size is 1 bin
angle_params = (-1, 1, angle_bin, -1, 1, angle_bin)  # angular binning for tan_x and tan_y
dx = scene_shape[0] / det_nx / 2
location_params = (dx, scene_shape[0] - dx, det_nx, dx, scene_shape[0] - dx, det_nx, -dx)  # min & max x&y locations and number of detectors
# alternatively location_params can be passed as a tuple containing (x,y,z) of every detector

# # defining the voxels
start = datetime.now()
print('Simulation of {}x{}x{} voxel model with {}x{} grid of detectors, minimum {} of intersecting detectors \n'.format(*scene_shape,det_nx,det_nx,multi_detect))
volo_init, volo_true = create_default_voxel_setting(scene_shape, delta_alpha, shift=0, size=size)
list_init = voxel_array_to_list(volo_init, size=size)
list_true = voxel_array_to_list(volo_true, size=size)
detectors, voxel_crosses_list = create_detectors(angle_params, location_params, list_true, list_init, size=size,
                                                 detector_shift=0)
print('initialization time:', datetime.now()-start)
```

#### Saving initial model and detector histograms

```python
if not os.path.exists(input_dir): os.mkdir(input_dir)
if os.path.exists(input_dir+'/3d_output'): shutil.rmtree(input_dir+'/3d_output')
os.mkdir(input_dir+'/3d_output')
save_voxel_model(volo_init, format='.dat', file_name=input_dir+'/3d_output/Model_initial')
save_voxel_model(volo_true, format='.dat', file_name=input_dir+'/3d_output/Model_true')
if os.path.exists(input_dir+'/detectors'): shutil.rmtree(input_dir+'/detectors')
os.mkdir(input_dir+'/detectors')
for i_d, det in detectors.items():
    save_detector(input_dir+'/detectors/det_init.txt', det, i_d)
```

#### Visualising one of the initial detectors

```python
i_d = 1; visualise_detector(detectors[i_d], i_det=i_d, iterated=False)
```

## Gradient descent iteration
* number of steps and learning rate are just empirically chosen in order for the result not to diverge.
* learning rate decay every n steps (by factor of 2) is implemented in the grad_step function

```python
start = datetime.now()
num_steps = 50
lr = 0.003 / det_nx / scale
list_pred = np.copy(list_init)
for step in range(num_steps):
    detectors, list_pred = grad_step(step, lr, detectors, list_pred, voxel_crosses_list, multi_det=multi_detect,
                                     loss_function='l2', n_decay=10, verbose=False)
volo_pred = voxel_list_to_array(list_pred, size=size)
print('iteration time:', datetime.now()-start)
```

#### Saving final model and detector histograms

```python
if not os.path.exists(output_dir): os.mkdir(output_dir)
if os.path.exists(output_dir + '/3d_output'): shutil.rmtree(output_dir + '/3d_output')
os.mkdir(output_dir + '/3d_output')
save_voxel_model(volo_pred, format='.dat', file_name=output_dir+'/3d_output/Model_simulated')
if os.path.exists(output_dir + '/detectors'): shutil.rmtree(output_dir + '/detectors')
os.mkdir(output_dir + '/detectors')
for i_d, det in detectors.items():
    save_detector(output_dir+'/detectors/det_final.txt', det, i_d, sim_types=['pred'])
```

#### Visualising the same detector after iterations

```python
i_d = 1; visualise_detector(detectors[i_d], i_det=i_d, iterated=True)
```

## Visualising the 3D models

#### True voxel model with highlighted unknown structures

```python
delta_vis = 0.8 # color of unknown structures is different for better visibility
volo_true_vis = np.copy(volo_init)
volo_true_vis += delta_vis * np.where(np.abs(volo_true-volo_init), delta_vis, 0)
visualise_voxels(volo_true_vis, inv=False, figsize=(5, 5))
```

#### Predicted voxel model, darker and more transparent voxel correspond to lower anomaly value (less empty)

```python
visualise_voxels(volo_pred, inv=False, figsize=(5,5), det_coords=np.array([d['coord'] for d in detectors.values()]))
```

#### Finding the optimal anomaly threshold
* In the experimental application threshold has to be chosen without the knowledge of the ground-truth 3D model, for instance the one with the best performance on toy models
* in the series of tests (in the 'tests' directory) the best performance was mostly achieved with a threshold 0.35-0.4

```python
thr_range = np.arange(0.25, 0.76, 0.05)
rve_range = [rve_score(volo_true, volo_init, volo_pred > th) for th in thr_range]
i_best = np.argmin(rve_range); thresh = thr_range[i_best]
print('Initial difference (in voxels): {:.2f}'.format(np.abs(volo_true-volo_init).sum()))
print('Best score:\n\t{:.2f} with threshold {:.2f}'.format(rve_range[i_best], thr_range[i_best]))
```

##### Predicted model after applying the threshold

```python
vox_pred_thresh = np.where(volo_pred>thresh, delta_alpha, 1.0 - delta_alpha)
visualise_voxels(vox_pred_thresh, inv=False, figsize=(5,5), det_coords=np.array([d['coord'] for d in detectors.values()]))
```

### Wrong voxels in the output after threshold

```python
visualise_voxels((np.abs(volo_true - vox_pred_thresh)).clip(0, 1), inv=False, figsize=(5, 5))
```
